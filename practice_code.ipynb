```python
pi = 3.14159
radius = 2.2
# calculates area of a circle (with radius "radius") and prints it
area = pi*(radius**2)
print(area)
```

```python
radius = radius + 1    # radius value is changed
print(area)            # no change to area value
area = pi*(radius**2)  # area value is recalculated based on new radius
print(area)            # new area value is printed
```
